```python
#Rounds always upwards to the nearest 5 (or base)

def round_up_specific(x, base=5):
    return (base* np.ceil(x/base))
```

```python


def MJO_Exdind_Heatmap(index, ex_type, save_fig = 0):

    data = MJO_Exinds[index][ex_type]
    data = data.replace(np.nan, 0)



    '''Setting Up DAta and TItles'''

    #Getting the data into the correct fornat\
    #Must be trasposed so that the phases are the column titles and the stations are the row titles
    data = data.transpose()
    data_np = np.array(data)

    #The titles that will be used for the heat map
    column_titles = data.columns
    row_titles= data.index


    #Mins and Maxs for the color bars to work of

    vmin  = np.min(data_np)
    vmax = np.max(data_np)



    '''Creating the Plot'''



    fig, ax  = plt.subplots(figsize = (20,20))
    plt.title(index + ' ' + ex_type + ' for Phases of the MJO', size = 15)

    # Creating the Color Mapping
    if ex_type == 'Frequency':
        delta = 5
        
    else:
        delta = 10
        
        
    rounded_up_max = round_up_specific(vmax, delta)
    num_boxes = rounded_up_max/delta
        
    cmap = cm.get_cmap('Blues', num_boxes)




    #The Actual Plot
    im = ax.imshow(data, vmin = 0 - delta, vmax = vmax + delta, cmap = cmap)



    '''X and Y Ticks'''
    ax.set_xticks(np.arange(len(column_titles)));
    ax.set_xticklabels(column_titles);

    ax.set_yticks(np.arange(len(row_titles)));
    ax.set_yticklabels(row_titles);



    '''Color Bar'''
    cbartick_labels = np.arange(0,rounded_up_max + delta, delta)
#     cbartick_labels = np.ceil(np.linspace(0, vmax, 10))


#     cbar = fig.colorbar(im, shrink = 0.45, ticks = cbartick_labels)
    cbar = fig.colorbar(im, shrink = 0.45, ticks = cbartick_labels, boundaries = cbartick_labels)
    cbar.ax.set_yticklabels(cbartick_labels);
    cbar.ax.set_title(index + ' ' + ex_type, fontsize = 10)

    if save_fig:
        fig.savefig(index + ' ' + ex_type + ' for Phases of the MJO.png', bbox_inches = 'tight')
```

```python
index_list = list(MJO_Exinds.keys())
ex_list = ['Frequency','Intensity (mm)']

for index in index_list[:3]:
    for ex_type in ex_list[:3]:

        MJO_Exdind_Heatmap(index, ex_type, save_fig = 1)
```

## North to South 

```python


def MJO_Exdind_Heatmap_N2S(index, ex_type, save_fig = 0):

    data = MJO_Exinds[index][ex_type].transpose()
    data = data.replace(np.nan, 0)


    data = data.reindex(MJO_df_N2S.index)
    

    '''Setting Up DAta and TItles'''

    #Getting the data into the correct fornat\
    #Must be trasposed so that the phases are the column titles and the stations are the row titles

    data_np = np.array(data)

    #The titles that will be used for the heat map
    column_titles = data.columns
    row_titles= data.index


    #Mins and Maxs for the color bars to work of

    vmin  = np.min(data_np)
    vmax = np.max(data_np)



    '''Creating the Plot'''



    fig, ax  = plt.subplots(figsize = (20,20))
    plt.title(index + ' ' + ex_type + ' for Phases of the MJO (N2S)', size = 15)

    # Creating the Color Mapping
    num_boxes = 10
    cmap = cm.get_cmap('Blues', num_boxes )
#     cmap = 'Blues'


    delta = 1
    # vmax/num_boxes/2


    #The Actual Plot
    im = ax.matshow(data, vmin = 0 - delta, vmax = vmax + delta, cmap = cmap)



    '''X and Y Ticks'''
    ax.set_xticks(np.arange(len(column_titles)));
    ax.set_xticklabels(column_titles);

    ax.set_yticks(np.arange(len(row_titles)));
    ax.set_yticklabels(row_titles);



    '''Color Bar'''

    cbartick_labels = np.ceil(np.linspace(0, vmax, 10))

    cbar = fig.colorbar(im, shrink = 0.45, ticks = cbartick_labels)
    cbar.ax.set_yticklabels(cbartick_labels);
    cbar.ax.set_title(index + ' ' + ex_type, fontsize = 10)

    if save_fig:
        fig.savefig(index + ' ' + ex_type + ' for Phases of the MJO (N2S).png', bbox_inches = 'tight')
```

```python
index_list = list(MJO_Exinds.keys())
ex_list = ['Frequency','Intensity (mm)']

for index in index_list:
    for ex_type in ex_list:

        MJO_Exdind_Heatmap_N2S(index, ex_type, save_fig = 1)
```

# Bar Plots

```python
index_list = ['', 'RX1d','P90','P95','P99']
index_names = ['Rainfall','Monthly Maximum Rainfall', 
               'Rainfall > 90th Percentile', 'Rainfall > 95th Percentile', 'Rainfall > 99th Percentile']

index_zip = list(zip(index_list , index_names))

ex_list = ['Frequency','Intensity (mm)']


#All fonts are adjusted from this
font_size = 12
row_col_size = font_size + 15

#Number of rows and columsn
ncols = 2
nrows= len(index_list)



fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(12, 15))

title = 'Rainfall Indices for Phases of the MJO'

plt.suptitle(title, size  = font_size + 10)



#This is so in order to get the correct title from index names list for each plot
row_num = 0


# This is to make sure that the other dict for normal rainfall gets plotted first
plot_num = 1

for row_num, index in enumerate(index_list): #Looping thought the rows
    

    for col_num,ex_type in enumerate(ex_list):
        
        
        '''Determining which data to use'''
        if row_num  == 0:
            if col_num == 0:
                data = MJO_normal_rain_dict['Frequency']
            elif plot_num  == 1:
                data = MJO_normal_rain_dict['Intensity (mm)']
        else:
            data = MJO_Exinds[index][ex_type]
        

        '''Calculations using the data'''
        #Finding the mean of each column
        average_of_phases = data.mean(axis = 1)
        
        
        #Find the Percent for Each Phase
        total_of_phase = np.array(average_of_phases.iloc[:].sum())
        prob_in_phase = average_of_phases / total_of_phase * 100
        
        
        
        
        ''' Plotting'''
        alpha = 1
        ax1 = axes[row_num, col_num]
        ax2 = ax1.twinx()
        ax1.bar(average_of_phases.index, average_of_phases.iloc[:], alpha = alpha )
        ax2.bar(average_of_phases.index,prob_in_phase, alpha =alpha )
        
        
        
        '''Y Limits'''
        #The upper limit on the y2 axis
        if ex_type == 'Frequency':
            percent_max = 22
        else:
            percent_max = 18
        
        y2_min, y2_max = ax2.get_ylim()
        
        delta_y = percent_max /y2_max #This is the amount the the y2 axis has been multiplided by in order to get
        #to percent_max
        
        
        y1_min, y1_max = ax1.get_ylim() #The current limits on the axis
        
        y1_max_new = y1_max * delta_y #The new limits on the axis, after it has been stretched by the same
        #amount
        
        
        ax1.set_ylim(0,y1_max_new)
        ax2.set_ylim(0,percent_max)
        
        
        '''Y labels'''
        if col_num == 0:
            ax1.set_ylabel('Average Frequency', size = 10)
        elif col_num == 1:
            ax1.set_ylabel('Average Intensity (mm)', size = 10)
            
        ax2.set_ylabel('Percent (%)', size = 10)

'''Row and Columns Titles'''
#Source of Code
# https://stackoverflow.com/questions/25812255/row-and-column-headers-in-matplotlibs-subplots


#Annotate Documenation
# https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.annotate.html


rows = index_zip
size = 14
pad = 5
for ax, row in zip(axes[:,0], rows):
    if row[1] == 'Rainfall': 
        ax.annotate(row[1], xy=(0, 0.5), xytext=(-ax.yaxis.labelpad - pad, 0),
                    xycoords=ax.yaxis.label, textcoords='offset points',
                    size=size, ha='right', va='center')
    else:
        ax.annotate(row[1] + '\n'+ '('+row[0] +')', xy=(0, 0.5), xytext=(-ax.yaxis.labelpad - pad, 0),
            xycoords=ax.yaxis.label, textcoords='offset points',
            size=size, ha='right', va='center')

    
cols = ex_list
    
for ax, col in zip(axes[0], cols):
    ax.annotate('Average ' + col, xy=(0.5, 1), xytext=(0, pad),
                xycoords='axes fraction', textcoords='offset points',
                size=size + 5, ha='center', va='baseline')

    
#Source:
# https://stackoverflow.com/questions/8248467/matplotlib-tight-layout-doesnt-take-into-account-figure-suptitle

fig.tight_layout(rect=[0, 0.03, 1, 0.92])


fig.savefig(title + '.png', bbox_inches = 'tight')
```
