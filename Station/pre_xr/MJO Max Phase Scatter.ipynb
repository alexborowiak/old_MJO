# Phase where most Extremes Occur

```python
#Referencing part of the dictionary so that it is easier to call in furutre
P90_freq = MJO_Exinds['P90']['Frequency']
P90_int = MJO_Exinds['P90']['Intensity (mm)']

P95_freq = MJO_Exinds['P95']['Frequency']
P95_int = MJO_Exinds['P95']['Intensity (mm)']


P99_freq = MJO_Exinds['P99']['Frequency']
P99_int = MJO_Exinds['P99']['Intensity (mm)']

RX1d_freq = MJO_Exinds['RX1d']['Frequency']
RX1d_int = MJO_Exinds['RX1d']['Intensity (mm)']

#Lists that will be added to the data frame
P90_freq_max = []
P90_freq_phase = []


P90_int_max = []
P90_int_phase = []



P95_freq_max = []
P95_freq_phase = []

P95_int_max = []
P95_int_phase = []

P99_freq_max = []
P99_freq_phase = []

P99_int_max = []
P99_int_phase = []

RX1d_freq_max = []
RX1d_freq_phase = []

RX1d_int_max = []
RX1d_int_phase = []

#station that will be used:
current_stations = P90_freq.columns

for station in current_stations:
        
    '''P90 Frequency'''
    #Calculating the maximum frequency
    maximum_freq_val_90 = P90_freq[station].max()
    P90_freq_max.append(maximum_freq_val_90)
    
    #The phase where the maximum frequency occurs
    phase_freq_val_90= P90_freq[station].idxmax()
    P90_freq_phase.append(phase_freq_val_90)
    
    
    '''P90 Intensity (mm)'''
    #Calculating the maximum average Intensity (mm)
    maximum_int_val_90 = P90_int[station].max()
    P90_int_max.append(maximum_int_val_90)
    
    #The phase where the maximum frequency occurs
    phase_int_val_90 = P90_int[station].idxmax()
    P90_int_phase.append(phase_int_val_90)
    ################
    
            
    '''P95 Frequency'''
    #Calculating the maximum frequency
    maximum_freq_val_95 = P95_freq[station].max()
    P95_freq_max.append(maximum_freq_val_95)
    
    #The phase where the maximum frequency occurs
    phase_freq_val_95= P95_freq[station].idxmax()
    P95_freq_phase.append(phase_freq_val_95)
    
    
    '''P95 Intensity (mm)'''
    #Calculating the maximum average Intensity (mm)
    maximum_int_val_95 = P95_int[station].max()
    P95_int_max.append(maximum_int_val_95)
    
    #The phase where the maximum frequency occurs
    phase_int_val_95 = P95_int[station].idxmax()
    P95_int_phase.append(phase_int_val_95)
    
    
    #################
    
    '''P99 Frequency'''
    #Calculating the maximum frequency
    maximum_freq_val_99 = P99_freq[station].max()
    P99_freq_max.append(maximum_freq_val_99)
    
    #The phase where the maximum frequency occurs
    phase_freq_val_99= P99_freq[station].idxmax()
    P99_freq_phase.append(phase_freq_val_99)
    
    
    '''P99 Intensity (mm)'''
    #Calculating the maximum average Intensity (mm)
    maximum_int_val_99 = P99_int[station].max()
    P99_int_max.append(maximum_int_val_99)
    
    #The phase where the maximum frequency occurs
    phase_int_val_99 = P99_int[station].idxmax()
    P99_int_phase.append(phase_int_val_99)
    
    #################
    
    
    
    
    '''RX1d Frequency'''
    #Calculating the maximum frequency
    RX1d_max_freq_val = RX1d_freq[station].max()
    RX1d_freq_max.append(RX1d_max_freq_val)
    
    #The phase where the maximum frequency occurs
    RX1d_phase_freq_val= RX1d_freq[station].idxmax()
    RX1d_freq_phase.append(RX1d_phase_freq_val)
    
    '''RX1d Intensity (mm)'''
    #Calculating the maximum intuency
    RX1d_max_int_val = RX1d_int[station].max()
    RX1d_int_max.append(RX1d_max_int_val)
    
    #The phase where the maximum intuency occurs
    RX1d_phase_int_val= RX1d_int[station].idxmax()
    RX1d_int_phase.append(RX1d_phase_int_val)

    
'''P90 Data Frame''' 


P90_df = pd.DataFrame({'Station': current_stations,
                       'Greatest Frequency': P90_freq_max,'Frequency Phase':P90_freq_phase,
                       'Greatest Intensity (mm)': P90_int_max,'Intensity (mm) Phase':P90_int_phase
                      })
P90_df = P90_df.set_index('Station')


'''P95 Data Frame''' 


P95_df = pd.DataFrame({'Station': current_stations,
                       'Greatest Frequency': P95_freq_max,'Frequency Phase':P95_freq_phase,
                       'Greatest Intensity (mm)': P95_int_max,'Intensity (mm) Phase':P95_int_phase
                      })
P95_df = P95_df.set_index('Station')


'''P99 Data Frame''' 


P99_df = pd.DataFrame({'Station': current_stations,
                       'Greatest Frequency': P99_freq_max,'Frequency Phase':P99_freq_phase,
                       'Greatest Intensity (mm)': P99_int_max,'Intensity (mm) Phase':P99_int_phase
                      })
P99_df = P99_df.set_index('Station')



'''RX1d Df'''


RX1d_df = pd.DataFrame({'Station':current_stations,
                        'Greatest Frequency': RX1d_freq_max,'Frequency Phase': RX1d_freq_phase,
                        'Greatest Intensity (mm)': RX1d_int_max,'Intensity (mm) Phase':RX1d_int_phase
                       })
RX1d_df = RX1d_df.set_index('Station')
```

```python
Max_Phase_df= pd.concat({'RX1d': RX1d_df, 'P90':P90_df, 'P95':P95_df ,'P99': P99_df }, axis = 1)
Max_Phase_df.head()
```

# Plotting 

```python
#Some useful examples
# https://scitools.org.uk/cartopy/docs/v0.13/matplotlib/gridliner.html

def max_phase_plot(index, ex_type, save_fig = 0):
    fig = plt.figure(figsize=(20,30))
    ax = plt.axes(projection=ccrs.PlateCarree())
    ax.coastlines(resolution='10m', linewidth = 1)
    plt.title("Phase of Max " + index + ' ' + ex_type, size = 30)



    ax.set_extent([110,157,-7.5,-22.7])


    current_stations = MJO_df.index
    colors = ['w','g','r','c','m','y','k','b']


    for station in current_stations:
        #Latitude and Longitude of Point
        lon = MJO_df['Lon'].loc[station]
        lat =  MJO_df['Lat'].loc[station]


        
        
        '''Variate vs Invariate Point Size'''
    
    
        '''Variate'''
        #The var will be size of the point, whilst the phase will be the color
#         var = Max_Phase_df[index]['Greatest ' + ex_type].loc[station]


#         #Getting the right size for marker points dependent on data type
#         if ex_type == 'Frequency':
#             var = var * 10
#         else:
#             var = var * 5

        '''Invariate '''
        var = 150
        phase = Max_Phase_df[index][ex_type + ' Phase'].loc[station]




        #Different points will have a different color depending on the phase that they occured at
        phase_color = colors[phase - 1]

        #Scatter the points onto the figure
        ax.scatter(float(lon), float(lat),transform=ccrs.PlateCarree(), s= var, c=phase_color)    

    '''Creating the Legend'''


    #Random points that don't appera to be used for the legend
    ####  Colors


    point_size = 150

    #Only want the phases that actually appeared to be in the scatter plot
    used_phases =  Max_Phase_df[index][ex_type + ' Phase'].unique()
    used_phases = np.sort(used_phases) #Sorting the phases so that they appear in the correct order in the legend


    for phase in used_phases:
        ax.scatter(112,-24,zorder = -100,s = point_size, c = colors[phase - 1], label = 'Phase ' + str(phase))


    ax.legend(loc = 'upper left', fontsize = 15)

    if save_fig:
        fig.savefig("Phase of Max " + index + ' ' + ex_type + '.png', bbox_inches = 'tight')
```

```python
#Going through all of the different exinds and both frequency and intensity plots are produced


index_list= ['RX1d','P90', 'P95','P99']
index = index_list[1]
ex_list = ['Frequency','Intensity (mm)']
ex_type = ex_list[0]


for index in index_list:
    for ex_type in ex_list:
        max_phase_plot(index, ex_type, save_fig = 1)
```
