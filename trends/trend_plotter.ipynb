## Map Trend Plotter: Single

This plots just the trends for a single type of rainfall (90, 95, all). This is the most up to date

```python
def trend_plot_single(data,cmap, vmax,vmin,title = '',save_fig = 0, dont_plot = 0):

    '''~~~~~~~~~~~ Plot Set-up'''

    height = 10
    fig = plt.figure(figsize = (15,height))
    title_size = 15
    row_nums = 4
    fontsize = 15
    fig.suptitle(title, fontsize = 20)

    plot_num = 1
    ############################# Plots
    '''~~~~~~~~~~~ Plot Loop'''
    # Looping through all the phase: 'enhanced', 'suppressed','all'. 
    # These will be the rows
    for phase in data.mjo.values:
        sub_data = data.sel(mjo = phase)

        ax = fig.add_subplot(row_nums,1,plot_num, projection = ccrs.PlateCarree())
        plot = sub_data.plot(ax = ax,vmax = vmax,vmin = vmin, cmap = cmap, add_colorbar = False)
        ax.set_title('')
        ax.set_extent((110,153,-22,-10.2))
        ax.outline_patch.set_visible(False)
        ax.coastlines(resolution = '50m')
        '''Seperators x-direction'''
        ydip = -0.1 # This is the value required to move the sepreatro down to the next plot
        ax.plot([0, 1], [ydip, ydip], color='black', lw=1, transform=ax.transAxes, clip_on=False)

        plot_num += 1   

    '''~~~~~~~~~~~ Colorbar'''
    # x,y, width, heigth
    cax = fig.add_axes([.75, 0.27,.02,0.3]) 
    step = 5
    boundaries  = np.arange(-vmax, vmax + step, step)
    cbar = plt.colorbar(plot, orientation = 'vertical', cax = cax,
                       extend = 'both', boundaries = boundaries)
    cbar.set_label('Percent Per \n Decade', fontsize = 12, labelpad = 32, rotation = 0)    
    
    '''~~~~~~~~~~~~~~~  Row Labels'''
    x_pos = 0.165
    ax.annotate('Enhanced', rotation = 90,xy=(x_pos, 0.85), fontsize = fontsize, 
        xycoords='figure fraction', textcoords='offset points',
        horizontalalignment='left', verticalalignment='top')
    ax.annotate('Suppressed', rotation = 90,xy=(x_pos, 0.62), fontsize = fontsize, 
        xycoords='figure fraction', textcoords='offset points',
        horizontalalignment='left', verticalalignment='top')
    ax.annotate('Inactive', rotation = 90,xy=(x_pos, 0.38), fontsize = fontsize, 
        xycoords='figure fraction', textcoords='offset points',
        horizontalalignment='left', verticalalignment='top')
    ax.annotate('All', rotation = 90,xy=(x_pos, 0.1), fontsize = fontsize, 
        xycoords='figure fraction', textcoords='offset points',
        horizontalalignment='left', verticalalignment='top')
    

    '''~~~~~~~~~~~ Saving'''
    fig.tight_layout(rect=[0, 0.03, 1, 0.93])

    if save_fig:
        fig.savefig(save_dir + title + '.png',bbox_inches = 'tight', dpi = 400)
        print('---' + title +  ' : has been saved')

    # This will not plot usefull if you just want to save
    if dont_plot:
        plt.close(fig)
```

## Map Trend Plotter: Multiple

This plots just the trends for a single type of rainfall (90, 95, all). This is the most up to date

```python
def trend_plot_multiple(data,percentile_list,cmap, vmax,vmin,title = '',save_fig = 0, dont_plot = 0):
    
    'Set-up'
    # This is how many collumns are going to be needed for the trend lis
    num_columns = len(percentile_list)
    num_rows = 4
    # Getting the right dimensions for both plots

    height = 12
    fig = plt.figure(figsize = (15,height))
    title_size = 15
    fontsize = 15
    fig.suptitle(title, fontsize = 20)

    plot_num = 1
    ############################# Plots
    
    # Looping through all the phase: 'enhanced', 'suppressed','all'. 
    # These will be the rows
    
    # Looping through the phases of the percentiles first, as plot number goes across rather than down 
    # (staying on enhanced, then going to suppressed)
    for phase in data.mjo.values:
         for percentile in percentile_list:
        

            sub_data = data.sel(mjo = phase, precip = percentile)

            ax = fig.add_subplot(num_rows,num_columns,plot_num, projection = ccrs.PlateCarree())
            plot = sub_data.plot(ax = ax,vmax = vmax,vmin = vmin, cmap = cmap, add_colorbar = False)
            
            
            ax.set_extent([110,157, -7.5,-23]) # Lat and long doesn't vary if there are no sig poitns on boudnary

            ax.coastlines(resolution = '50m')
            
            
            if plot_num <= num_columns: # This ensures just the top row
                # This ensure the names are being plotted correclty
                if percentile == 'all':
                    col_title = 'All Rainfall'
                else:
                    col_title = str(percentile) + 'th Percentile'
                
                plt.title(col_title, size = fontsize)
            else:
                plt.title('')
                


            plot_num += 1   

    '''------ Color Bar'''
    
    
#     ticks= np.array([0.4,0.6,0.8,1,1.2,1.4,1.6,1.8,2])
#     tick_labels = [str(i) for i in ticks]
#     ticks = ticks - 0.05
#     boundaries = np.arange(0.1,vmax - step,step)
    
    # This addes in a seperate axis for the color bar: x,y,width, height
    cax = fig.add_axes([.3, 0.9, 0.4,0.028])
#     boundaries  = np.linspace(-vmax, vmax, 11)
    step = 5
    boundaries  = np.arange(-vmax, vmax + step, step)
    cbar = plt.colorbar(plot, orientation = 'horizontal', cax = cax,
                       extend = 'both', boundaries = boundaries)
#     ticks = np.array(boundaries).astype(str)
#     ticks = [i + '%' for i in ticks]
#     cbar.ax.tick_params(labelsize = 10)
#     cbar.ax.set_yticklabels(ticks) 
    cbar.set_label('Percent Per Decade', fontsize = 12, labelpad = 4, rotation = 0)    
    
    '''Row Labels'''
    yheight = 4.6
    for i in range(num_rows):
        
        title = data.mjo.values[i].capitalize()
        if title == 'All':
            title = 'All Phases'
        else:
            title = title
        
        plt.text(-2.25, yheight, title, horizontalalignment='center', verticalalignment='center',
                     transform=ax.transAxes, fontsize = fontsize)
        yheight -= 1.365

    
    
    
    #######
    fig.tight_layout(rect=[0, 0.03, 1, 0.89])

#####################
    #####
    if save_fig:
        fig.savefig(save_dir + title + '('+','.join(percentile_list.astype(str)) + ').png',bbox_inches = 'tight', dpi = 300)
        print('---' + title +  ' : has been saved')
        
        
        
    ####
    # This will not plot usefull if you just want to save
    if dont_plot:
        plt.close(fig)
```
